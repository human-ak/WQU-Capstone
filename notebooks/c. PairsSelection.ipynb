```python
series_analyser = class_SeriesAnalyser.SeriesAnalyser()
trader = class_Trader.Trader()
data_processor = class_DataProcessor.DataProcessor()
```

```python
# split data in training and test
df_prices_train, df_prices_test = data_processor.split_data(df_prices,
                                                            ('01-01-2018',
                                                             '31-12-2023'),
                                                            ('01-01-2024',
                                                             '31-12-2024'),
                                                            remove_nan=True)
train_val_split = '01-01-2023'
```

# Pairs Selection

```python
subsample = 2500
min_half_life = 21 # avg. number of points in a month
max_half_life = 252 # avg. number of points in a year
```

## PCA pairs selection

```python
import gdown

# import csv file with PCA clusters
pca_clusters_url = "https://drive.google.com/uc?id=1NERf_6JdL2DPLipSPLzcakrXRFj-RS4q"
output_file = "pca_clusters.csv"
gdown.download(pca_clusters_url, output_file, quiet=False)

# load in a df
pca_clusters = pd.read_csv(output_file)
pca_clusters.rename(columns={"Index": "Ticker", "Cluster": "Category_cluster"}, inplace = True)
pca_clusters.set_index("Ticker", inplace = True)
```

```python
pairs_category, unique_tickers = series_analyser.get_candidate_pairs(
                                                            clustered_series = pca_clusters,
                                                            pricing_df_train=df_prices_train,
                                                            pricing_df_test=df_prices_test,
                                                            min_half_life=min_half_life,
                                                            max_half_life=max_half_life,
                                                            min_zero_crosings=12,
                                                            p_value_threshold=0.01,
                                                            hurst_threshold=0.5,
                                                            subsample=subsample,
                                                            )

pairs_found_category = len(pairs_category)

pairs_found_category
```

```python
import pickle
with open('PCA_selected_pairs.pickle', 'wb') as f:
   pickle.dump(pairs_category, f)
```
